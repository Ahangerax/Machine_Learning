```python
import seaborn as sns 

df = sns.load_dataset('tips')

df.head()
```

```python
# label encoding in sckit learn of tips dataset
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['day'])
df['day_encoded'] = le.transform(df['day'])

df.head()
```

```python
# reverse encoding of day column
df['day_decoded'] = le.inverse_transform(df['day_encoded']) 
df.head()
```

```python
# feature scaling using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(df[['total_bill', 'tip']])
df[['total_bill_scaled', 'tip_scaled']] = scaled_values
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df[['total_bill','size']]
y = df['tip']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

y_pred = model.predict(x_test)

print(mean_squared_error(y_pred,y_test))
```
