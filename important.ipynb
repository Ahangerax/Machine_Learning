# NumPy is used for fast numerical operations, arrays, matrices, and math functions.
## It is the backbone of Pandas, Matplotlib, ML, and Deep Learning.

--

```python
import numpy as np

# -------------------------------
# Creating NumPy Arrays
# -------------------------------

# 1D array
arr = np.array([1, 2, 3, 4, 5])

# 2D array
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])

# âŒ Python list math is slow
# âœ” NumPy arrays are fast and vectorized


# -------------------------------
# Array Properties
# -------------------------------
arr.shape        # Shape of array
arr.dtype        # Data type
arr.size         # Total number of elements


# -------------------------------
# Mathematical Operations
# -------------------------------

# âŒ Loop-based math (slow)
# âœ” Vectorized operations (fast)
squared = arr ** 2
added = arr + 10


# -------------------------------
# Useful Array Creation
# -------------------------------
zeros = np.zeros(5)          # [0. 0. 0. 0. 0.]
ones = np.ones(5)            # [1. 1. 1. 1. 1.]
range_arr = np.arange(0, 10, 2)  # [0 2 4 6 8]
linspace_arr = np.linspace(0, 1, 5)


# -------------------------------
# Statistics
# -------------------------------
arr.mean()
arr.sum()
arr.max()
arr.min()


# -------------------------------
# Reshaping Arrays
# -------------------------------
reshaped = arr.reshape(5, 1)

# âŒ Wrong reshape causes error
# âœ” Total elements must match
```

#ğŸ“˜ PANDAS NOTES (Data Handling & Analysis)
##ğŸ”¹ What is Pandas?

###Pandas is used for structured data (tables like Excel / SQL).
----

```python
import pandas as pd

# -------------------------------
# Creating DataFrame
# -------------------------------
data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 25, 30],
    "City": ["Delhi", "Mumbai", "Chennai"]
}

df = pd.DataFrame(data)


# -------------------------------
# Viewing Data
# -------------------------------
df.head()        # First 5 rows
df.tail()        # Last 5 rows


# -------------------------------
# Selecting Columns
# -------------------------------
names = df["Name"]        # Single column
subset = df[["Name", "Age"]]  # Multiple columns


# -------------------------------
# Filtering Rows
# -------------------------------
young = df[df["Age"] > 25]

# âŒ df[df[int"Age" < 20]]
# âœ” df["Age"] < 20


# -------------------------------
# Adding New Column
# -------------------------------
df["Salary"] = [30000, 40000, 50000]

# âŒ df.append("Salary")  (wrong)
# âœ” Direct column assignment


# -------------------------------
# Sorting
# -------------------------------
df_sorted = df.sort_values(by="Age", ascending=False)


# -------------------------------
# Statistics
# -------------------------------
avg_age = df["Age"].mean()


# -------------------------------
# Rename Column
# -------------------------------
df.rename(columns={"City": "Location"}, inplace=True)

# âŒ replace() changes values, not column names


# -------------------------------
# Missing Values
# -------------------------------
df.isnull().sum()


# -------------------------------
# Drop Missing Values
# -------------------------------
df_clean = df.dropna(subset=["Age"])


# -------------------------------
# Read CSV File
# -------------------------------
df_csv = pd.read_csv("./data.csv")
```

#ğŸ“˜ MATPLOTLIB NOTES (Data Visualization)

##ğŸ”¹ What is Matplotlib?

###Matplotlib is used for visualizing data using plots and graphs.

```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------
# Sample Data
# -------------------------------
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 30, 40])


# -------------------------------
# Line + Scatter Plot
# -------------------------------
plt.figure(figsize=(6, 4))

plt.plot(x, y, label="Line", marker='o')
plt.scatter(x, y)

# Labels and title
plt.title("Relation between Time and Distance")
plt.xlabel("Time")
plt.ylabel("Distance")

# Grid and legend
plt.grid(True)
plt.legend()

plt.show()

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.grid(True)
plt.show()

```
